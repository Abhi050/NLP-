```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
## Tokenization --- converts paragraph into sentences and focus on the words
nltk.download("punkt")
sentences = nltk.sent_tokenize(paragraph)
```

```python
## Steming

stemmer = PorterStemmer()

stemmer.stem("thnking harding")

```

```python
## Lemmatization

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

lemmatizer.lemmatize("goes")
```

```python
## Cleaning the data in paragraph

import re
corpus = []

for i in range(len(sentences)):
  review = re.sub("[^a-zA-Z]"," ", sentences[i])
  review = review.lower()
  corpus.append(review)

corpus
```

```python
 nltk.download('stopwords')
stopwords.words('english')
```

```python
for j in corpus:
  words = nltk.word_tokenize(j)
  for word in words:
    print(stemmer.stem(word))
```

```python
## Applyig Stemming with stopords

for j in corpus:
  words = nltk.word_tokenize(j)
  for word in words:
    if words not in stopwords.words("english"):
      print(stemmer.stem(word))
```

```python
## Applying Lemmatization
for j in corpus:
  words = nltk.word_tokenize(j)
  for word in words:
    if words not in stopwords.words("english"):
      print(lemmatizer.lemmatize(word))
```

```python
## Bag of Words

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X = cv.fit_transform(corpus)

# Binary bagging
cv2 = CountVectorizer(binary = True)
X2 = cv2.fit_transform(corpus)
```

```python
## apply Stopwords. Lemmatize 

import re 
corpus1 =[]
for i in range(len(sentences)):
  review2= re.sub("[^a-zA-Z]"," ",sentences[i])
  review2 = review2.lower()
  review2 = review2.split()
  review2 = [lemmatizer.lemmatize(word) for word in review2 if not word in stopwords.words("english")]
  review2 = " ".join(review2)
  corpus1.append(review2)
```
